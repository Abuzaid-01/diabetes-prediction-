```python
file['BloodPressure'] = file['BloodPressure'].replace(0, file['Insulin'].mean())
file['SkinThickness'] = file['SkinThickness'].replace(0, file['SkinThickness'].median())
file['Insulin'] = file['Insulin'].replace(0, file['Insulin'].median())

```

```python
## Splitting the data into input features (X) and target value (y)
X = file.drop(columns='Outcome', axis=1)
y = file['Outcome']


```

```python
cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']
for x in cols:
  Q1 = X[x].quantile(0.25)
  Q3 = X[x].quantile(0.75)
  IQR = Q3 - Q1
  lower = Q1 - 1.5 * IQR
  upper = Q3 + 1.5 * IQR
  mask = (X[x] >= lower) & (X[x] <= upper)
```

```python
x_outlier = X[mask]
y_outlier = y[mask]
```

```python
file.shape
```

```python
x_outlier.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x_outlier)
```

```python
X_scaled = pd.DataFrame(X_scaled, columns=x_outlier.columns)
X_scaled.reset_index(drop=True, inplace=True)
y_outlier = pd.Series(y_outlier) # or pd.DataFrame
y_outlier.reset_index(drop=True, inplace=True)
```

```python
X_scaled.shape
```

```python
q = X_scaled['Insulin'].quantile(.95)
mask = X_scaled['Insulin'] < q
dataNew = X_scaled[mask]
y_outlier = y_outlier[mask]
```

```python
fig, ax = plt.subplots(figsize=(15,15))
sns.boxplot(data = dataNew, ax = ax)
plt.savefig("boxplot2.jpg")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataNew, y_outlier, test_size=0.3, stratify=y_outlier, random_state=2)
```

```python
y_train.value_counts()
```

```python

from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Check resampled class distribution
print("\nResampled class distribution:")
print(pd.Series(y_train_resampled).value_counts())
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic.fit(X_train,
             y_train)

logistic2 = LogisticRegression()
logistic2.fit(X_train_resampled,
             y_train_resampled)
```

```python
y_predictions = logistic.predict(X_test)
print(y_predictions)

y_predictionss = logistic2.predict(X_test)
print(y_predictionss)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predictions)

from sklearn.metrics import classification_report
target_names = ['Non-Diabetic', 'Diabetic']
print(classification_report(y_test, y_predictions, target_names=target_names))


from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predictionss)

from sklearn.metrics import classification_report
target_names = ['Non-Diabetic', 'Diabetic']
print(classification_report(y_test, y_predictionss, target_names=target_names))
```

```python
accuracy = accuracy_score(y_test, y_predictions)
accuracyy = accuracy_score(y_test, y_predictionss)

```

now we apply ensemble technique


```python
train, val_train, test, val_test = train_test_split(dataNew, y_outlier, test_size=0.5, stratify=y_outlier, random_state=20)
X_train, X_test, y_train, y_test = train_test_split(train, test, test_size=0.2, random_state=355)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(X_train, y_train)
kn.score(X_test, y_test)
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
svc.score(X_test, y_test)
```

```python
col1 = kn.predict(val_train)
col2 = svc.predict(val_train)
column = np.column_stack((col1, col2))
column
```

```python
predict_test1 = kn.predict(X_test)

predict_test2 = svc.predict(X_test)

predict_test = np.column_stack((predict_test1, predict_test2))

print(predict_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(column, val_test)
rf.score(predict_test, y_test)
```

```python
grid_param = {
    "n_estimators": [90, 100, 115],
    'criterion': ['gini', 'entropy'],
    'min_samples_leaf': [1, 2, 3, 4, 5],
    'min_samples_split': [4, 5, 6, 7, 8],
    'max_features': ['auto', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator=rf, param_grid=grid_param, cv=5, n_jobs=-1, verbose=3)
grid_search.fit(column, val_test)
```

```python
best_params_=grid_search.best_params_
```

```python
best = RandomForestClassifier(**best_params_)
best.fit(column, val_test)
best.score(predict_test, y_test)
```
